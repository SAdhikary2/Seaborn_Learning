# Learning Seaborn 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
mtcar=pd.read_csv('mtcars.csv')
```

```python
mtcar.head()
```

```python
mtcar.info()
```

```python
mtcar.shape
```

```python
#bar plot is not working due to barplot can not take more than one argument

res=sns.barplot(mtcar['cyl'],mtcar['mpg'])

plt.show()

```

```python
res=sns.barplot(x='cyl', y='mpg', data=mtcar)
#plt.show()
res
```

```python
#this is using the matplotlib

res1=plt.bar(mtcar['cyl'],mtcar['mpg'])
plt.show()
```

```python
res=sns.barplot(x='cyl', y='mpg', data=mtcar,color='yellow')
plt.show()
```

```python
# COUNT PLOT

# sns.countplot(x='cyl',data=mtcar,palette='Set1')


sns.countplot(x='cyl',data=mtcar)
```

```python
sns.countplot(x='gear',hue='cyl',data=mtcar,palette='Set1')
```

```python
#DISTPLOT

sns.distplot(mtcar.mpg,bins=5,color='g')
```
